# Determinization

Function `treeAutDeterminization()` takes 2 inputs:
- a `TTreeAut` object
- an `alphabet`, which is structured as a dictionary: to each symbol from the alphabet, its arity is assigned

The function not only determinizes the automaton, it also creates a complete TA, that is why the `alphabet` is required.
* the function works in exponential time, so with bigger automata, it may take longer
* the state labeled `{}` is basically the "sink" state, needed for completeness of the automaton

```python
H0box = boxesDict["boxH0"]
convertToDOT(H0box)
```

```python
H0boxDet = treeAutDeterminization(H0box, {"LH": 2, "0": 0, "1": 0, "Port_H0":0})
convertToDOT(H0boxDet)
```

# Union

- the function `treeAutUnion()` takes two TAs as an input an returns an automaton that generates the union of the two languages generated by the input TAs
- all states with the same names from between the two TAs are renamed such that no name colissions (and thus structural collisions) occur
- the renaming happens before the union creation

```python
L0box = boxesDict["boxL0"]
print(matchTreeBU(L0box, testTreeDict["treeL0test1"]))
print(matchTreeBU(L0box, testTreeDict["treeL0test2"]))
print(matchTreeBU(L0box, testTreeDict["treeL0test3"]))
print(matchTreeBU(L0box, testTreeDict["treeL0test4"]))
print(matchTreeBU(L0box, testTreeDict["treeL1test1"]))
print(matchTreeBU(L0box, testTreeDict["treeL1test2"]))
print(matchTreeBU(L0box, testTreeDict["treeL1test3"]))
print(matchTreeBU(L0box, testTreeDict["treeL1test4"]))
convertToDOT(L0box)
```

```python
L1box = boxesDict["boxL1"]
print(matchTreeBU(L1box, testTreeDict["treeL0test1"]))
print(matchTreeBU(L1box, testTreeDict["treeL0test2"]))
print(matchTreeBU(L1box, testTreeDict["treeL0test3"]))
print(matchTreeBU(L1box, testTreeDict["treeL0test4"]))
print(matchTreeBU(L1box, testTreeDict["treeL1test1"]))
print(matchTreeBU(L1box, testTreeDict["treeL1test2"]))
print(matchTreeBU(L1box, testTreeDict["treeL1test3"]))
print(matchTreeBU(L1box, testTreeDict["treeL1test4"]))
convertToDOT(L1box)
```

```python
L0L1union = treeAutUnion(L0box, L1box)
print(matchTreeBU(L0L1union, testTreeDict["treeL0test1"]))
print(matchTreeBU(L0L1union, testTreeDict["treeL0test2"]))
print(matchTreeBU(L0L1union, testTreeDict["treeL0test3"]))
print(matchTreeBU(L0L1union, testTreeDict["treeL0test4"]))
print(matchTreeBU(L0L1union, testTreeDict["treeL1test1"]))
print(matchTreeBU(L0L1union, testTreeDict["treeL1test2"]))
print(matchTreeBU(L0L1union, testTreeDict["treeL1test3"]))
print(matchTreeBU(L0L1union, testTreeDict["treeL1test4"]))
convertToDOT(L0L1union)
```

# Intersection

- the function `treeAutIntersection()` takes two TAs and returns their intersection
- the states inside the intersection have 'tuple names', e.g. `'(q1,q2)'` (`q1` is from TA1, `q2` is from TA2)

```python
H1box = boxesDict["boxH1"]
convertToDOT(H1box)
```

Even just by looking at the structure of boxes H0 and H1, we can tell that their intersection will not produce any trees, because their output symbols are completely different.

```python
H0H1intersection = treeAutIntersection(H0box, H1box)
convertToDOT(H0H1intersection)
```

# Complement

- the function `treeAutComplement()` takes two inputs: a TA and an alphabet
- before the complementation, a determinization on the TA takes place (hence the input alphabet)
- complementation itself is just invert operation on all the rootstates of the determinzed (and complete) TA

```python
Xbox = boxesDict["boxX"]
print(matchTreeBU(Xbox, testTreeDict["treeXtest1"]))
print(matchTreeBU(Xbox, testTreeDict["treeXtest2"]))
print(matchTreeBU(Xbox, testTreeDict["treeXtest3"]))
print(matchTreeBU(Xbox, testTreeDict["treeL0test1"]))
print(matchTreeBU(Xbox, testTreeDict["treeL1test2"]))
print(matchTreeBU(Xbox, testTreeDict["treeH0test3"]))
print(matchTreeBU(Xbox, testTreeDict["treeH1test4"]))
convertToDOT(Xbox)
```
