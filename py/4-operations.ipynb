# Basic operations with tree automata

In this notebook, we will show how to use the basic automata operations (namely union, intersection, complementation, determinization) from the module `ta_functions.py`.

```python
from tree_automata import (
    TTreeAut,
    tree_aut_union,
    tree_aut_intersection,
    tree_aut_determinization,
    tree_aut_complement,
    match_tree_bottom_up,
)
from helpers.utils import box_catalogue
import tests.test_data as td
from formats.render_dot import convert_to_dot
```

# Determinization

Function `tree_aut_determinization()` takes 2 inputs:
- a `TTreeAut` object
- an `alphabet`, which is structured as a dictionary: to each symbol from the alphabet, its arity is assigned

The function not only determinizes the automaton, it also creates a complete TA, that is why the `alphabet` is required.
* the function works in exponential time, so with bigger automata, it may take longer
* the state labeled `{}` is basically the "sink" state, needed for completeness of the automaton

```python
box_H0: TTreeAut = box_catalogue["boxH0"]
convert_to_dot(box_H0)
```

```python
box_H0_determinized: TTreeAut = tree_aut_determinization(box_H0, {"LH": 2, "0": 0, "1": 0, "Port_H0":0})
convert_to_dot(box_H0_determinized)
```

# Union

- the function `tree_aut_union()` takes two TAs as an input an returns an automaton that generates the union of the two languages generated by the input TAs
- all states with the same names from between the two TAs are renamed such that no name collisions (and thus structural collisions) occur
- the renaming happens before the union creation

```python
L0box: TTreeAut = box_catalogue["boxL0"]
print(match_tree_bottom_up(L0box, td.l0_test_tree_1))
print(match_tree_bottom_up(L0box, td.l0_test_tree_2))
print(match_tree_bottom_up(L0box, td.l0_test_tree_3))
print(match_tree_bottom_up(L0box, td.l0_test_tree_4))
print(match_tree_bottom_up(L0box, td.l1_test_tree_1))
print(match_tree_bottom_up(L0box, td.l1_test_tree_2))
print(match_tree_bottom_up(L0box, td.l1_test_tree_3))
print(match_tree_bottom_up(L0box, td.l1_test_tree_4))
convert_to_dot(L0box)
```

```python
L1box: TTreeAut = box_catalogue["boxL1"]
print(match_tree_bottom_up(L1box, td.l0_test_tree_1))
print(match_tree_bottom_up(L1box, td.l0_test_tree_2))
print(match_tree_bottom_up(L1box, td.l0_test_tree_3))
print(match_tree_bottom_up(L1box, td.l0_test_tree_4))
print(match_tree_bottom_up(L1box, td.l1_test_tree_1))
print(match_tree_bottom_up(L1box, td.l1_test_tree_2))
print(match_tree_bottom_up(L1box, td.l1_test_tree_3))
print(match_tree_bottom_up(L1box, td.l1_test_tree_4))
convert_to_dot(L1box)
```

```python
L0L1union: TTreeAut = tree_aut_union(L0box, L1box)
print(match_tree_bottom_up(L0L1union, td.l0_test_tree_1))
print(match_tree_bottom_up(L0L1union, td.l0_test_tree_2))
print(match_tree_bottom_up(L0L1union, td.l0_test_tree_3))
print(match_tree_bottom_up(L0L1union, td.l0_test_tree_4))
print(match_tree_bottom_up(L0L1union, td.l1_test_tree_1))
print(match_tree_bottom_up(L0L1union, td.l1_test_tree_2))
print(match_tree_bottom_up(L0L1union, td.l1_test_tree_3))
print(match_tree_bottom_up(L0L1union, td.l1_test_tree_4))
convert_to_dot(L0L1union)
```

# Intersection

- the function `tree_aut_intersection()` takes two TAs and returns their intersection
- the states inside the intersection have 'tuple names', e.g. `'(q1,q2)'` (`q1` is from TA1, `q2` is from TA2)

```python
box_H1: TTreeAut = box_catalogue["boxH1"]
convert_to_dot(box_H1)
```

Even just by looking at the structure of boxes H0 and H1, we can tell that their intersection will not produce any trees, because their output symbols are completely different.

```python
intersection_H0_H1: TTreeAut = tree_aut_intersection(box_H0, box_H1)
convert_to_dot(intersection_H0_H1)
```

# Complement

- the function `tree_aut_complement()` takes two inputs: a TA and an alphabet
- before the complementation, a determinization on the TA takes place (hence the input alphabet)
- complementation itself is just invert operation on all the rootstates of the determinzed (and complete) TA

```python
box_X: TTreeAut = box_catalogue["boxX"]
print(match_tree_bottom_up(box_X, td.x_test_tree_1))
print(match_tree_bottom_up(box_X, td.x_test_tree_2))
print(match_tree_bottom_up(box_X, td.x_test_tree_3))
print(match_tree_bottom_up(box_X, td.l0_test_tree_1))
print(match_tree_bottom_up(box_X, td.l1_test_tree_2))
print(match_tree_bottom_up(box_X, td.h0_test_tree_3))
print(match_tree_bottom_up(box_X, td.h1_test_tree_4))
convert_to_dot(box_X)
```

```python
box_X_complement: TTreeAut = tree_aut_complement(box_X, td.full_alphabet)
print(match_tree_bottom_up(box_X_complement, td.x_test_tree_1))
print(match_tree_bottom_up(box_X_complement, td.x_test_tree_2))
print(match_tree_bottom_up(box_X_complement, td.x_test_tree_3))
print(match_tree_bottom_up(box_X_complement, td.l0_test_tree_1))
print(match_tree_bottom_up(box_X_complement, td.l1_test_tree_2))
print(match_tree_bottom_up(box_X_complement, td.h0_test_tree_3))
print(match_tree_bottom_up(box_X_complement, td.h1_test_tree_4))
convert_to_dot(box_X_complement)
```
