# Testing emptiness of the TA language

- functions `nonEmptyBU()` (non-recursive) and `nonEmptyTD()` (recursive) take a tree automaton structure as an input and they return a tuple of two elements (TTreeNode, string):
    - the first element is a root node of a "witness" tree
    - the second element is a string representation of this "witness" tree
    * "witness" tree is a trivial tree that can be generated by the particular TA
    and serves as a proof of the non-emptiness
- if the language of the TA is empty, then a tuple (None, "") is returned.

```python
from ta_classes import * 
from ta_lib import * 

from test_data import *

from format_dot import *
from format_vtf import *
from format_tmb import *

from jupyter import *
```

```python
convertToDOT(boxesDict["boxL1"], 'a')
```

```python
witnessTree, witnessString = nonEmptyBU(boxesDict["boxL1"])

print(witnessString)
convertToDOT(witnessTree, 'n')
```

It is clear that box L1 does not have an empty language. However with using a `removeState()` method we can alter this box such that it will have an empty language. The witness tree will then be None and `convertToDOT()` will return an empty DOT image.

```python
tempBox = copy.deepcopy(boxesDict["boxL1"])
tempBox.removeState('s2')

convertToDOT(tempBox)
```

```python
testTree, testString = nonEmptyTD(tempBox)

if testString == "":
    print("testString is empty")
else:
    print(testString)
convertToDOT(testTree, 'n')
```
